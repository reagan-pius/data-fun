## 1. NumPy
- Stands for Numerical Python
- Behaves like python lists but is much faster
- Can be created from a python list:  
>**np.array([1, 4, 2, 5, 3])**  
>**np.array([1, 2, 3, 4], dtype='float32')***  
>###### - A list of lists in converted to a 2-dimensional array
>**original_list = [[1,2,3],[3,4,5],[6,7,8]]
two_dimensional_array = np.array(original_list)**

```python
# begin by importing it 
import numpy as np 

# create a NumPy array

# An array of length 10, filled with zero
np.zeros(10, dtype=int)

# An array of size 3x5 filled with 1.0
np.ones((3,5), dtype=float)

# An array of containing a linear sequence starting at 0 and going upto 20, steps of two
np.arange(0,20,2)

# An array of five numbers, linearly spaced between 0 and 1
np.linspace(0,1,5)

# An array of the given shape and populate it with random samples.
np.random.random((3,3))

# The identity matrix of size 3
np.eye(3)
```

### Some more hands on

```python
np.random.seed(0)
x1 = np.random.randint(10, size=6)

# 1D array
print('Number of dimensions: ', x1.ndim)
print('Shape: ', x1.shape)
print('size: ', x1.size)
print('Type: ', x1.dtype)

x2 = np.random.randint(10, size=(3, 4))
print(x2[0,1])

# Simple mathematics first
x = np.arange(4)
print("x     =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
# Divide and round
print("x // 2 =", x // 2)  

x = [-2, -1, 1, 2]

print("Absolute value: ", np.abs(x))
print("Exponential: ", np.exp(x))
print("Logarithm: ", np.log(np.abs(x)))

M = np.random.random((3, 4))
print(M)
# Note the syntax variable.function instead of 
# np.fonction(variable). Both are acceptable if
# variable is a Numpy array.
print("Sum of all elements of M: ", M.sum())
print("Sums of the columns of M: ", M.sum(axis=0))
```

## 2. Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

### A more professional plot

```python
# Change default font size
plt.rcParams.update({'font.size': 15})

fig = plt.figure()
ax = plt.axes()

# Solid line, color specified by its name
plt.plot(x, np.sin(x - 0), color='blue', linestyle='solid', label='bleu')
# Short name for color, dashed line
plt.plot(x, np.sin(x - 1), color='g', linestyle='dashed', label='vert')
# Grayscale between 0 and 1, dashes and dots
plt.plot(x, np.sin(x - 2), color='0.75', linestyle='dashdot', label='gris')
# RGB color, dotted line
plt.plot(x, np.sin(x - 3), color='#FF0000', linestyle='dotted', label='rouge')
# Axis limits. Try also 'tight' and 'equal' to see their effect
plt.axis([-1, 11, -1.5, 1.5]);
# Labels
plt.title("Example of a graph")
# The legend is generated from the argument 'label' of 'plot'
# 'loc' specified the placement of the legend.
plt.legend(loc='lower left');
# Axis titles
ax = ax.set(xlabel='x', ylabel='sin(x)')
```
